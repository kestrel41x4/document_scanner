```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import imutils
```

```python
img_orig = cv2.imread('./images/03.jpg')
```

```python
cv2.namedWindow('original',cv2.WINDOW_NORMAL)
cv2.imshow('original',img_orig)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
def resizer(image,width=500):
    h,w,c = image.shape
    
    height = int((h/w)* width )

    
    size = (width,height)
    image = cv2.resize(image,(width,height))
    return image, size
```

```python
img_re,size = resizer(img_orig)
print(size)
cv2.imshow('Resize',img_re)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
detail = cv2.detailEnhance(img_re,sigma_s = 20, sigma_r = 0.15)
gray = cv2.cvtColor(detail,cv2.COLOR_BGR2GRAY) # GRAYSCALE IMAGE
blur = cv2.GaussianBlur(gray,(5,5),0)
# edge detect
edge_image = cv2.Canny(blur,75,200)
# morphological transform
kernel = np.ones((5,5),np.uint8)
dilate = cv2.dilate(edge_image,kernel,iterations=1)

closing = cv2.morphologyEx(dilate,cv2.MORPH_CLOSE,kernel)

contours , hire = cv2.findContours(closing,
                                   cv2.RETR_LIST,
                                   cv2.CHAIN_APPROX_SIMPLE)

contours = sorted(contours, key=cv2.contourArea, reverse=True)
for contour in contours:
    peri = cv2.arcLength(contour, True)
    approx = cv2.approxPolyDP(contour,0.02*peri, True)

    if len(approx) == 4:
        fp = np.squeeze(approx)
        break

cv2.drawContours(img_re, [fp],-1,(0,255,0),3)


multiplier = img_orig.shape[1] / size[0]
fpo = fp * multiplier
fpo = fpo.astype(int)

```

```python
from imutils.perspective import four_point_transform
```

```python
warp_image = four_point_transform(img_orig,fpo)
```

```python
cv2.imshow('wrap', warp_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def document_scanner(image):

    img_re,size = resizer(image)


    detail = cv2.detailEnhance(img_re,sigma_s = 20, sigma_r = 0.15)
    gray = cv2.cvtColor(detail,cv2.COLOR_BGR2GRAY) # GRAYSCALE IMAGE
    blur = cv2.GaussianBlur(gray,(5,5),0)
    # edge detect
    edge_image = cv2.Canny(blur,75,200)
    # morphological transform
    kernel = np.ones((5,5),np.uint8)
    dilate = cv2.dilate(edge_image,kernel,iterations=1)

    closing = cv2.morphologyEx(dilate,cv2.MORPH_CLOSE,kernel)

    contours , hire = cv2.findContours(closing,
                                    cv2.RETR_LIST,
                                    cv2.CHAIN_APPROX_SIMPLE)

    contours = sorted(contours, key=cv2.contourArea, reverse=True)
    for contour in contours:
        peri = cv2.arcLength(contour, True)
        approx = cv2.approxPolyDP(contour,0.02*peri, True)

        if len(approx) == 4:
            fp = np.squeeze(approx)
            break

    cv2.drawContours(img_re, [fp],-1,(0,255,0),3)


    multiplier = image.shape[1] / size[0]
    fpo = fp * multiplier
    fpo = fpo.astype(int)

    warp_image = four_point_transform(image,fpo)

    return warp_image, fpo, img_re, closing
```

```python
img = cv2.imread('./images/1725235814.png')
wrpimg, points, cnt_img, edgeimg = document_scanner(img)
```

```python
cv2.imshow('img', img)
cv2.imshow('warp', cnt_img)
cv2.imshow('resize', edgeimg)
cv2.imshow('warp', wrpimg)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
